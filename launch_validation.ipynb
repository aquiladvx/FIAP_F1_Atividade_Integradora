# 4. Verificação para decolagem

```python
def check_launch(row):
    if row["structural_integrity"] != 1:
        return "ABORTED: structural failure"

    if row["energy_level"] < 80:
        return "ABORTED: low energy"

    if not (5 <= row["internal_temp"] <= 35):
        return "ABORTED: internal temperature out of range"

    if not (-50 <= row["external_temp"] <= 50):
        return "ABORTED: external temperature out of range"

    if not (200 <= row["tank_pressure"] <= 300):
        return "ABORTED: tank pressure out of range"

    if row["critical_modules"] != "OK":
        return "ABORTED: failure on critical modules"

    return "READY FOR LAUNCH"
```

# 5. Aplicação da função de verificação

```python
random_index = random.randint(0, len(df) - 1)
random_row = df.iloc[random_index]

result = check_launch(random_row)

print("Selected telemetry reading:")
print(random_row)
print("\nLaunch decision:")
print(result)
```
